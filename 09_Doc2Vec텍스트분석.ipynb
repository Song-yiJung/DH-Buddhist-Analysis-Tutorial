í™œìš© ë°©ì•ˆ: ì €ìë‚˜ ì—°ëŒ€ë¥¼ ì•Œ ìˆ˜ ì—†ëŠ” ìˆ˜ì²œ ê°œì˜ ê³ ë¬¸ì„œ ë­‰ì¹˜(ì˜ˆ: ê°œì¸ì´ ê¸°ì¦í•œ í¸ì§€ ëª¨ìŒ)ë¥¼ Doc2Vecìœ¼ë¡œ í•™ìŠµì‹œí‚¨ë‹¤. ê·¸ ê²°ê³¼ë¡œ ë‚˜ì˜¨ ë¬¸ì„œ ë²¡í„°ë“¤ì„ K-Meansì™€ ê°™ì€ êµ°ì§‘í™” ì•Œê³ ë¦¬ì¦˜ì— ì ìš©í•˜ë©´, ì»´í“¨í„°ê°€ ìë™ìœ¼ë¡œ ë¬¸ì„œë“¤ì„ ë‚´ìš©ì— ë”°ë¼ ì£¼ì œë³„(ì˜ˆ: 'ê°€ì¡± ì•ˆë¶€', 'ì •ì¹˜ ë™í–¥', 'ê¸ˆì „ ê±°ë˜')ë¡œ ë¶„ë¥˜í•´ì¤€ë‹¤. ì´ëŠ” ëŒ€ê·œëª¨ ì•„ì¹´ì´ë¸Œ ì •ë¦¬ ë° íƒìƒ‰ì˜ ì¶œë°œì ì´ ëœë‹¤.

```python
# 1. í•„ìˆ˜ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜
# - `pip install --no-deps` ì˜µì…˜ì„ ì‚¬ìš©í•˜ì—¬ gensimë§Œ ì„¤ì¹˜í•©ë‹ˆë‹¤.
# - ì´ë ‡ê²Œ í•˜ë©´ numpyì™€ scipyì˜ ì¶©ëŒì„ ë¬´ì‹œí•˜ê³  gensimì„ ì„¤ì¹˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
# - numpyì™€ scipyëŠ” Colabì— ì´ë¯¸ ì„¤ì¹˜ë˜ì–´ ìˆìœ¼ë¯€ë¡œ, gensimë§Œ ì„¤ì¹˜í•˜ë©´ ë©ë‹ˆë‹¤.
!pip install gensim --no-deps -q
print("âœ… í•„ìˆ˜ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜ ì™„ë£Œ: gensim\n")

# 2. ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸ ë° ë°ì´í„° ì¤€ë¹„
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import os

# ê°€ìƒì˜ ë””ì§€í„¸ ì¸ë¬¸í•™ ë¬¸ì„œ (í•œë¬¸ ë²•êµ¬ê²½ êµ¬ì ˆ)
documents = [
    "ç¡çœ  è§£å¯¤",
    "å®œ æ­¡å–œ æ€",
    "è½ æˆ‘ æ‰€ èªª",
    "æ’°è¨˜ ä½›è¨€",
    "è¡†ç”Ÿ ç›¸å‰‹",
    "è€æ­» çŒ¶ç„¶",
    "å‘½ çµ‚ è‡ªç„¶",
    "å¾ ç”Ÿæ­» å¾—åº¦",
    "ç„¡ ç• ç„¡ æ‚£",
    "è¡Œ ä¸ é›¢ å¿ƒ",
    "ä¸–é–“ è¬ç‰© ç„¡å¸¸",
    "å–„ è¨€ å®‰æ¨‚",
]

# 3. ë°ì´í„° ì „ì²˜ë¦¬ (í† í°í™” ë° Doc2Vec í¬ë§·ìœ¼ë¡œ ë³€í™˜)
tagged_data = [
    TaggedDocument(words=doc.split(), tags=[i])
    for i, doc in enumerate(documents)
]

print("âœ… ë°ì´í„° ì „ì²˜ë¦¬ ë° TaggedDocument ë³€í™˜ ì™„ë£Œ")
print(f"ë³€í™˜ëœ ë°ì´í„° ì˜ˆì‹œ (1ë²ˆ ë¬¸ì„œ): {tagged_data[1]}\n")

# 4. Doc2Vec ëª¨ë¸ í•™ìŠµ
model = Doc2Vec(vector_size=50, min_count=1, epochs=40)
model.build_vocab(tagged_data)
model.train(tagged_data, total_examples=model.corpus_count, epochs=model.epochs)

print("âœ… Doc2Vec ëª¨ë¸ í•™ìŠµ ì™„ë£Œ\n")


# 5. ëª¨ë¸ì„ í™œìš©í•œ ë¬¸ì„œ ìœ ì‚¬ë„ ë¶„ì„
# 'ë¬¸ì„œ 4'ë¥¼ ê¸°ì¤€ ë¬¸ì„œë¡œ ì‚¼ì•„, ì´ ë¬¸ì„œì™€ ê°€ì¥ ìœ ì‚¬í•œ ë¬¸ì„œë¥¼ ì°¾ìŠµë‹ˆë‹¤.
query_doc_id = 4
query_doc_vector = model.dv[query_doc_id]

similar_docs = model.dv.most_similar(positive=[query_doc_vector], topn=3)

print(f"ğŸ” ê¸°ì¤€ ë¬¸ì„œ ({query_doc_id}ë²ˆ ë¬¸ì„œ)ì™€ ê°€ì¥ ìœ ì‚¬í•œ ë¬¸ì„œ Top 3:\n")
print(f"--- ê¸°ì¤€ ë¬¸ì„œ ë‚´ìš© ---\n{documents[query_doc_id]}\n")

for doc_id, similarity in similar_docs:
    if doc_id == query_doc_id:
        continue
    print(f"----------------------------------------")
    print(f"ìœ ì‚¬ ë¬¸ì„œ ID: {doc_id}ë²ˆ ë¬¸ì„œ")
    print(f"ìœ ì‚¬ë„ ì ìˆ˜: {similarity:.4f}")
    print(f"ë¬¸ì„œ ë‚´ìš©: {documents[doc_id]}\n")

```

```python
# --- ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸° ---
import requests
from bs4 import BeautifulSoup
import jieba
from opencc import OpenCC
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import numpy as np
import re # ì •ê·œí‘œí˜„ì‹ ë¼ì´ë¸ŒëŸ¬ë¦¬

# --- 1. ë°ì´í„° ìë™ í™•ë³´ ë° í’ˆ(å“) ë‹¨ìœ„ ë¶„ë¦¬ ---
url = "https://raw.githubusercontent.com/cbeta-org/xml-p5/master/T/T04/T04n0210.xml"
documents = []
chapter_titles = []

try:
    response = requests.get(url)
    response.raise_for_status()

    # BeautifulSoupì´ CBETAì˜ ë„¤ì„ìŠ¤í˜ì´ìŠ¤(cb:)ë¥¼ ì¸ì‹í•˜ë„ë¡ 'lxml-xml' íŒŒì„œë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
    soup = BeautifulSoup(response.content, 'lxml-xml')

    # [í•µì‹¬ ìˆ˜ì •] CBETA P5 í‘œì¤€ì— ë”°ë¼, ê° í’ˆ(å“)ì„ ê°ì‹¸ëŠ” <cb:div type="pin"> íƒœê·¸ë¥¼ ì°¾ìŠµë‹ˆë‹¤.
    chapters = soup.find_all('cb:div', {'type': 'pin'})

    for chapter in chapters:
        # <cb:mulu> íƒœê·¸ì—ì„œ í’ˆ(å“)ì˜ ì œëª©ì„ ì¶”ì¶œí•©ë‹ˆë‹¤.
        title_tag = chapter.find('cb:mulu')
        title = title_tag.get_text(strip=True) if title_tag else "ì œëª© ì—†ìŒ"

        # í•´ë‹¹ í’ˆì— ì†í•œ ëª¨ë“  <p> íƒœê·¸ì˜ í…ìŠ¤íŠ¸ë¥¼ í•©ì³ ë³¸ë¬¸ì„ ë§Œë“­ë‹ˆë‹¤.
        text_parts = [p.get_text(strip=True) for p in chapter.find_all('p')]
        full_text = "".join(text_parts)

        if full_text:
            documents.append(full_text)
            chapter_titles.append(title)

    print(f"âœ… ã€ë²•êµ¬ê²½ã€ ì›ë¬¸ì—ì„œ ì´ {len(documents)}ê°œì˜ í’ˆ(ç« )ì„ ìë™ìœ¼ë¡œ ë¶ˆëŸ¬ì™”ìŠµë‹ˆë‹¤.")
    if documents:
        print(f"   - ì²« ë²ˆì§¸ í’ˆ: {chapter_titles[0]}")
        print(f"   - ë§ˆì§€ë§‰ í’ˆ: {chapter_titles[-1]}")
    else:
        print("\nâš ï¸ ê²½ê³ : XMLì—ì„œ í’ˆ(chapter) ë°ì´í„°ë¥¼ ì¶”ì¶œí•˜ì§€ ëª»í–ˆìŠµë‹ˆë‹¤. XML êµ¬ì¡°ë‚˜ URLì„ í™•ì¸í•´ì£¼ì„¸ìš”.")

except Exception as e:
    print(f"ì˜¤ë¥˜: ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ê±°ë‚˜ ì²˜ë¦¬í•˜ëŠ” ì¤‘ ë¬¸ì œê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. ({e})")


# --- 2. í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬ ë° TaggedDocument ì´í„°ë ˆì´í„° ìƒì„± ---
class TaggedDocumentIterator:
    def __init__(self, documents):
        self.documents = documents
        self.tokenizer = jieba.Tokenizer()
        self.cc = OpenCC('t2s')

    def __iter__(self):
        for i, doc in enumerate(self.documents):
            simplified_doc = self.cc.convert(doc)
            yield TaggedDocument(words=self.tokenizer.lcut(simplified_doc), tags=[i])

if documents:
    tagged_corpus = TaggedDocumentIterator(documents)
    print("\nâœ… ë°ì´í„° ì²˜ë¦¬ê¸°(Iterator) ì¤€ë¹„ ì™„ë£Œ.")

    # --- 3. Doc2Vec ëª¨ë¸ í›ˆë ¨ ---
    model = Doc2Vec(
        vector_size=100,
        window=5,
        min_count=2,
        workers=-1,
        dm=1,
        seed=42,
        epochs=50
    )

    model.build_vocab(tagged_corpus)
    print("âœ… ì–´íœ˜ ì‚¬ì „ êµ¬ì¶• ì™„ë£Œ.")

    model.train(tagged_corpus, total_examples=model.corpus_count, epochs=model.epochs)
    print("âœ… Doc2Vec ëª¨ë¸ í›ˆë ¨ ì™„ë£Œ!")

    # --- 4. ìœ ì‚¬ ë¬¸ì„œ íƒìƒ‰ ---
    try:
        # ì œëª©ì—ì„œ í’ˆ ë²ˆí˜¸ì™€ ì´ë¦„ì„ í•¨ê»˜ ê²€ìƒ‰í•˜ì—¬ ì •í™•ë„ë¥¼ ë†’ì…ë‹ˆë‹¤.
        query_title_keyword = 'ç„¡å¸¸å“'
        query_doc_id = next(i for i, title in enumerate(chapter_titles) if query_title_keyword in title)

        similar_docs = model.dv.most_similar(query_doc_id, topn=5)

        print(f"\n--- ğŸš€ ê¸°ì¤€ ë¬¸ì„œ [{query_doc_id}ë²ˆ: {chapter_titles[query_doc_id]}]ê³¼ ìœ ì‚¬í•œ í’ˆ(ç« ) Top 5 ---")
        for doc_id, similarity in similar_docs:
            print(f"  - ìœ ì‚¬ë„: {similarity:.4f} | [{doc_id}ë²ˆ í’ˆ] {chapter_titles[doc_id]}")
    except (ValueError, StopIteration):
        print(f"ì˜¤ë¥˜: ê¸°ì¤€ ë¬¸ì„œ '{query_title_keyword}'ì„(ë¥¼) ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
```
